```python
%matplotlib inline
import openmc
import math

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.ticker as tk 
import numpy as np
```

```python
# List of reactors to study 
reactors=['Ref_PWR', 'Micro_PWR', 'HTGR', 'Heat_pipe', 'FCM-HTGR']
```

```python
# Select the name of the metrics
metrics=['activity', 'activity_conv', 'decay_heat', 'decay_heat_conv', 'radiotoxicity', 'radiotoxicity_conv']
metrics_prop=['activity_prop', 'decay_heat_prop','radiotoxicity_prop' ] # list with the metric with the relative contribution of each nuclide
metrics_name=['Activity (Bq/g)', 'activity (Bq/GWe.y)', 'decay_heat (W/g)', 'decay_heat (W/GWe.y)', 'radiotoxicity (Sv/g)', 'radiotoxicity (Sv/GWe.y)'] # legend of the graph

```

```python
# Return a list of dataframe "metrics" associated to each reactor

def load_results(reactors, metrics):
    results=[]
    for i in reactors: # iterate on each reactor
        result=pd.read_csv('/home/cchristiaen/Results/'+i+'/'+metrics+'_'+i+'.csv', sep=',', ) # create a dataframe with the csv file
        isotope=result.iloc[:, 0]
        result.set_index(isotope, inplace=True) # Set radionuclide name as index
        result_list=result.iloc[:, 1:].copy(deep=True)
        results.append(result_list)
    return results
```

```python
# create a list of list with dataframe with metrics results
def load_metrics(reactors, metrics): # metrics is a list with the metrics name
    results=[]
    for i in metrics:
        result=load_results(reactors, i)
        results.append(result)
    return results
```

```python
# Create a dictionnary with the discharge time corresponding to each reactor
reactor_list=[ 'Ref_PWR', 'Micro_PWR','HTGR_20', 'HTGR_40','HTGR_60','HTGR', 'HTGR_100','FCM-HTGR', 'Heat_pipe', 'SFR', 'test_SFR_2']
discharge_time=[1273,560,396, 792, 1191,1589,1986, 3849, 1658, 9917, 9917]
dict_time=dict(zip(reactor_list, discharge_time))
    
```

```python
results=load_metrics(reactors, metrics)
results_prop=load_metrics(reactors, metrics_prop)
print (results_prop)
```

```python
# Function that select the time index the closest to the time of interest 'time value'
def select_time(result, time_value, reactor): # time value must be given in days
    time_value=time_value*365+dict_time[reactor]
    res=result.transpose()
    time=res.index.tolist()
    time=[float(x) for x in time]
    res['time']=time
    idx= res['time'].sub(time_value).abs().idxmin()
    return idx
```

```python
# Sort each column of the dataframe and select the 'number' first values 
# Return a dataframe with the 'number' radionuclides contributing most to the metrics at each time step of the simulation

def sort_df(result, number):
    df_sorted=pd.DataFrame()
    for column in result:
        df=result[column].sort_values(axis=0, ascending=False).head(number)
        df_sorted=pd.concat([df_sorted, df], axis=1)
    return df_sorted
    
```

```python
# return a list of sorted dataframes for each metric and each reactor
def sort_df_reactors(results, number):
    sort_results=[]
    for i in results:
        df=sort_df(i, number)
        sort_results.append(df)
    return sort_results
```

```python
number=10 # number of radionuclide contributing the most to the metrics
time_value=100 # time of interest for the metric for ex. activity at 10y
```

```python
# Plot most contributing radionuclide to the metric, for each reactor and each metric

for i in range (len(results)): # iterate on the list of metrics of interest
    fig, ax = plt.subplots()
    sort_results=sort_df_reactors(results[i], number) # create a list of sorted dataframe
    
    df=pd.DataFrame() # dataframe containing the "number" radionuclides contributing most to the metrics for each reactor

    for j in range (len(reactors)): # iterate on the list of reactors
    
        idx=select_time(results[i][j], time_value, reactors[j]) # select the time of interest for the metrics
        df_sorted=sort_results[j][idx].transpose().dropna() # select for the reactor 'j' the most contributinf nuclide at the time of interest
        df=pd.concat([df, df_sorted], axis=1) # 
    

    df.plot.bar(ax=ax, ylabel=metrics_name[i], xlabel='Nuclide', title=str(number)+' isotopes contributing most to the '+ metrics_name[i]+' at ' +str(time_value)+' years' )
    ax.legend(reactors)
    ax.set_yscale('log')
    print(metrics[i])
    display(df)
```

# Contribution of each nuclide to the waste metric, averaged on the selection of reactors 

The aim of this section is to compute the averaged contribution of each nuclide for the metrics of interest. For example, one result could be 'On average, 70% contribution to the activity at 100, 000 years is from Pu239'

```python
# Return a dataframe with the averaged relative contribution to each nuclide at the time of interest
## Results_metrics_prop corresponds to a list specific to a metric, containing for each reactor a dataframe with the relative contribution of each nuclide for the metric

def prop_total(reactors, results_metrics_prop, time, number):
    prop_reactors=pd.DataFrame() # Dataframe which columns corresponds to the metrics per nuclide at the time of interest 
    for i in range (len(reactors)): # iterate on the reactors of the study
        idx=select_time(results_metrics_prop[i], time, reactors[i]) # select the time of interest for ex. 10y
        df=results_metrics_prop[i].loc[:, idx].copy(deep=True) # select the column of the dataframe with the detailed value at this time
        prop_reactors[i]=df
    prop_tot=prop_reactors.sum(axis=1)/len(reactors) # dataframe with the averaged contribution of each nuclide to the metric 
    prop_tot=prop_tot.sort_values(axis=0, ascending=False) # sort the dataframe in descending order
    if (number!=None): # select the 'number' most contributing nuclides
        prop_tot=prop_tot.head(number)
    return prop_tot
    
        
```

```python
# Return a dataframe with the averaged relative contribution of each nuclide for each metric
## Results correspond to list containing for each metric a list of the radionuclide relative contribution specific for each reactor
def prop_metrics(reactors, results, time, number):
    prop_results=pd.DataFrame()
    columns_name=[]
    for i in range (len(results)): # iterate on the list of metrcis
        for j in time: # iterate on the list of time of interest 
            df=prop_total(reactors, results[i], j, number)
            prop_results=pd.concat([prop_results, df], axis=1, join='outer')
            columns_name.append(metrics_prop[i]+ ' '+str(j))
    prop_results.columns=columns_name
    return prop_results
```

```python
df_metrics=prop_metrics(reactors, results_prop, [10, 100, 100000], 5)
df_metrics=df_metrics.fillna(0)
df_metrics_report=df_metrics.round(decimals=2)
df_metrics_report
```
