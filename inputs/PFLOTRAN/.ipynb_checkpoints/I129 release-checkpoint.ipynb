## I-129 Inventory
This section extracts the concentration of I129 in the spent fuel from the "cooling_reactor.csv" files. With the parameters of the canister, it is possible to calculate the quantity of I129 in each canister.

```python
# Load the cooling dataframe results from OpenMC simulation
def get_reactor_results(reactors):
    results=[]
    for i in reactors:
        result=pd.read_csv('./Results/'+i+'/cooling'+i+'.csv', sep=',') # path to post treatment results
        results.append(result)
    return results
```

```python
# Define the parameters of the canister for each reactor, based on the design obtained in the repository footprint calculation that enables to get the lower total area
# Return a dictionnary with the HM_loading per fuel unit, the number of fuel unit and the number of canister in the reference repository
def get_canister_param(repository_capacity,reactor):
    canister={}
    if (reactor=='Ref_PWR'):
        canister={'HM_loading_per_unit': 460000, '# fuel units': 2 }
    elif (reactor=='Micro_PWR'):
        canister={'HM_loading_per_unit': 460000, '# fuel units': 12}
    elif (reactor=='HTGR'):
        canister={'HM_loading_per_unit': 8600, '# fuel units': 78}
    elif (reactor=='Heat_pipe'):
        canister={'HM_loading': 8600, '# fuel unit': 78}
    elif (reactor=='FCM-HTGR_per_unit'):
        canister={'HM_loading_per_unit': 21200, '# fuel units': 18}
    else:
        canister={'HM_loading_per_unit': 98450, '# fuel units': 3} # corresponds to the SFR reactor
    nCanisters = repository_capacity/(canister['HM_loading_per_unit']*canister['# fuel units'])
    canister['# canisters'] = nCanisters
    return canister 
```

```python
# Return the mass of I129 in the reactor specific canister. 
def get_I129_canister(repository_capacity, reactor, canister, result):
    # I129 parameters 
    half_life  = 15.7*365*24*60*60*(10**6) # in seconds
    Na         = 6.022*(10**23) # Avogadro
    molar_mass = 129 #g/mol
    
    canister_loading = canister['HM_loading_per_unit']*canister['# fuel units'] # mass of heavy metal in the canister
    I129_prop        = result['I129 (atom/cm3)'][0]*molar_mass/Na*result['volume_Ueq'][0]/result['mass U'][0]
    I129_mass        = I129_prop*canister_loading # mass if I129 per canister
    return I129_mass
```

## Fuel matrix dissolution

The degradation rate of the matrix in the environment will depend on the characteristics of the matrix. In this study, it is supposed that the degradation rate of the matrix is constant. The matrix contained in one canister is modelled by a sphere of equivalent mass: this geometrical trasnformation enables to easily consider the evolution of the matrix mass and volume during the degradation process. The radius of the sphere is given by:

$r_{0}=\frac{3M}{4\rho\pi}^{1/3}$

Then, considering a constant degradation rate of the matrix in the environment, the evolution of the radius with time is given as: $ r(t)=r_{0} - R*t$ with $R$ the degradation rate of the matrix. Then, the fraction of the matrix remaining in the canister is simply obtained with: $ f(t)= (1-\frac {Rt}{r_{0}\rho})^3 $

```python
def get_matrix_param(reactor):
    """
    Define the characteristic of the matrix for each reactor
    Return a dictionary with the name of the matrix, the mass of matrix per fuel unit, the low and high degradation rate, and the matrix density
    """
    matrix = {}
    if ('PWR' in reactor):
        matrix = {'name': 'UO2',     'mass per fuel unit': 521848.7395, 'low degradation rate': 2.2e-03, 'high degradation rate': 1.1e-02, 'density': 10.5, 'IRF': 0.01}
    elif (reactor=='Micro_PWR'):
        matrix = {'name': 'UO2',     'mass per fuel unit': 521848.7395, 'low degradation rate': 2.2e-03, 'high degradation rate': 1.1e-02, 'density': 10.5, 'IRF': 0.01}
    elif ('HTGR' in reactor):
        matrix = {'name': 'Graphite','mass per fuel unit': 117849.9979, 'low degradation rate': 7.4e-08, 'high degradation rate': 1.3e-06, 'density':1.85, 'IRF': 0.01}
    elif (reactor=='Heat_pipe'):
        matrix = {'name': 'Graphite','mass per fuel unit': 117849.9979, 'low degradation rate': 7.4e-08, 'high degradation rate': 1.3e-06, 'density': 1.85, 'IRF': 0.01}
    elif (reactor=='FCM-HTGR'):
        matrix = {'name': 'Graphite','mass per fuel unit': 95492.24603, 'low degradation rate': 7.4e-08, 'high degradation rate': 1.3e-06, 'density': 1.85, 'IRF': 0.01}
    else:
        matrix = {'name': 'Metallic','mass per fuel unit': 109388.8889, 'low degradation rate': 1.57,    'high degradation rate': 3.46,    'density':15.8, 'IRF': 0.01}
    return matrix 
```

```python
def get_eq_radius(reactor,matrix,tot_mass):
    """
    Calculate the equivalent radius of the canister before canister failure 
    """
    density  = matrix['density'] # density of the matrix
    radius   = (3*tot_mass/(4*np.pi*density))**(1/3) # equivalent radius for a matrix sphere with the same mass
    return radius
```

```python
def fraction_matrix(reactor, canister, level, t):
    """
    Calculate the fraction of the matrix remaining in the canister at time t in years
    - level:'high' or 'low' degradation rate
    - t : time (list) of interest after canister failure, can be a float or a list of float
    """
    
    # Canister and matrix parameters
    matrix   = get_matrix_param(reactor)
    tot_mass = matrix['mass per fuel unit']*canister['# fuel units']
 
    radius   = get_eq_radius(reactor,matrix,tot_mass)
    degradation_rate =matrix[level+' degradation rate']
    density  = matrix['density']
    
    fraction=(1-degradation_rate/(density*radius)*(10**-4)*t*365)**3
    fraction[fraction < 0] = 0
    return fraction
```

```python
def df_fraction_matrix(repository_capacity, reactors, level, t):
    """
    Create a dataframe with the fraction of the matrix remaining in the waste package
    """
    df=pd.DataFrame()
    for i in range (len(reactors)):
        reactor  = reactors[i]
        canister = get_canister_param(repository_capacity,reactor)
        f_matrix = fraction_matrix(reactor, canister,level, t)
        df[reactor] = f_matrix
    df.index = t
    return df
```

# I129 release in the environment

The conservative behaviour of I129 in the environment enables to make several hypothesis:
- The concentration of I129 in the spent fuel is constant (no decay)
- I129 is released instantly at canister failure $(2,5%)$ and the rest is released congruently to the waste matrix, at the same rate.

Therefore, the total I129 released in the environment after $t$ is given as: $M_{I129}(t)=M_{I129}(0)(IRF+(1-IRF)(1-(1-\frac {Rt}{r_{0}\rho})^3))$

```python
# Calculate the total quantity of I129 (in mol) released in the environment at time t in years for one canister
def get_I129_release_can(reactor, result, level, t):
    I129_rep = get_I129_canister(repository_capacity, reactor, canister, result)/molar_mass # I129 inventory in 1 canister in mol
    matrix_f = fraction_matrix(reactor, canister, level, t) # fraction of the matrix remaining in the waste package
    IRF      = get_matrix_param(reactor)['IRF']
    
    if (reactor in ['HTGR', 'Heat_pipe', 'FCM-HTGR']): # test if the reactor uses TRISO fuel, in which case the instant release fraction is 0
        I129_release=I129_rep*(1-matrix_f)
    else:
        I129_release=I129_rep*(IRF+(1-IRF)*(1-matrix_f))
    return I129_release
```

```python
# Calculate the total quantity of I129 released in the environment for a repository with a capacity of 70,000 MTU

def tot_I129_env_rep(reactor, result, level, t):
    nber_canister = canister(reactor)['Number of canister']
    return tot_I129_env_can(reactor, result, level, t)*nber_canister # the quantity released per canister is multiply by the number of canister corresponding to the repository capacity
```

```python
# Create a dataframe with the values of I129 released in the environment for a repository containing 70,000 MTU at different time

def I129_release_reactors(time, results, level, reactors):
    df=pd.DataFrame() # Dataframe containing the quantity of I129 released in the environment for all the reactors
    for i in range (len(reactors)):
        reactor = reactors[i]
        result  = results[i]
        I129_rep = get_I129_canister(repository_capacity, reactor, result)
        I129_reactor=tot_I129_env_rep(reactor, result, level, time) # evolution of the quantity of I129 released by one reactor
        df[reactor]=I129_reactor
    df.index=time
    return df
```

```python
# Create a dataframe with the values of I129 released in the environment for one canister at different time

def I129_release_reactors_can(time, results, level, reactors):
    df=pd.DataFrame()
    for i in range (len(reactors)):
        reactor=reactors[i]
        result=results[i]
        I129_reactor=tot_I129_env_can(reactor, result, level, time)
        df[reactor]=I129_reactor
    df.index=time
    return df
```

# Parameters of the study
This part calculates the total quantity of I129 released in the environment with the paramaters with the same methodoly as presented above from the study P.E Mariner, Application of generic Disposal System Models, 2015. 

```python
# Parameters of the study P. Mariner

Repository_capacity=70000*10**6 #g spent fuel -> the repository is considered to be full
I129_concentration_study=1.69*10**-6 # moles per gram UNF
IRF=0.025 # instant release fraction considered in the study
degradation_rate=4.8e-08 # mol/m2.s
degradation_rate=1.0 # gU/m2.day
N_can_study=23000
SNF_burnup=60 #GWd/MTU
```

```python
# Calculate the total quantity of I129 (in mol per MTU) released in the environment at time t in P.E Mariner

def I129_release_model(t):
    tot_mass=Repository_capacity*((238+32)/238) # total matrix mass in the repository
    tot_mass_can=tot_mass/N_can_study # total matrix mass in one canister
    radius=(3*tot_mass_can/(4*np.pi*10.5))**(1/3) # calculate the equivalent radius for the study 
    fraction=(1-degradation_rate/(10.5*radius)*(10**-4)*t*365)**3 # calculate the fraction of the matrix remaining in the waste package
    #I129_rep=I129_study*Repository_capacity
    I129_release=I129_concentration_study*(IRF+(1-IRF)*(1-fraction)) # mol of I129 per MTU
    return I129_release
```

```python
def I129_release_rate4PFLOTRAN(I129_release):

    for reactor in I129_rep_high.columns:
        time = I129_release.index
        
        r = I129_rep_high[reactor]
        t = np.diff(time)
        r = np.diff(r)

        time = time[1:]
        rate = r/t
        time = time[::2]
        rate = rate[::2]
        cons = rate/1e-6

        with open('PFLOTRAN/CONSTRAINT_'+reactor+'.txt','w') as file:
            for i in range(len(time)):
                file.write('CONSTRAINT cons'+str(i)+'\n')
                file.write('  CONCENTRATIONS \n')
                file.write('    I129   {cons:.2e}   T\n'.format(cons = cons[i]))
                file.write('  /\n')
                file.write('END\n')
        file.close()
    
        with open('PFLOTRAN/CONSTRAINT_LIST.txt','w') as file:
            file.write('0.d0 initial \n')
            for i in range(len(time)):
                file.write('{time:.2e}'.format(time=time[i])+'  cons'+str(i)+'\n')
        file.close()
   
```
