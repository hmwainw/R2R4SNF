## 1. Introduction

An advanced fuel cycle can reduce the inventory of actinides requiring permanent disposal but typically also generates multiple waste forms with higher heat loads due to the concentration of relatively short-lived fission products. In a generic repository with direct contact emplacement, the temperature of the engineered and/or natural barriers must be maintained below a certain limit to avoid compromising their performance. In order to comply with such limits three strategies can be implemented:

1. reduce the amount of waste loaded in a waste package; 
2. increase the spacing in between waste packages; 
3. increase the surface storage time. 

The goal of this study is to explore the relationship between advanced fuel cycles and waste management strategies, and in particular, on the required repository footprint versus surface storage time and waste package loading. This work considers two advanced fuel cycles: 
1. a limited-recycle fuel cycle in which uranium and plutonium recovered from light water reactors (LWRs) used nuclear fuel (UNF) are recycled in LWRs as mixed-oxide fuel (MOX); 
2. a continuous-recycle fuel cycle in which transuranics (TRU) from LWR UNF are used as makeup fuel for sodium cooled fast reactors (SFRs). 

In addition, waste from an LWR once-through fuel cycle is examined as the reference case. The repository temperature constraints are evaluated using analytical solutions for conductive heat transfer to inform the required spacing between waste packages. Waste volumes are normalized to the electricity generated. The corresponding repository footprint is determined as a function of waste form, waste package loading, surface storage time, and host rock. Three generic repository geological environments – granite, clay, and salt – are considered.

In this notebook, I will:
* describe the calculation to determine the number of canisters required to dispose of wastes from three fuel cycles;
* analyze the data regarding the area required per package of each of the wastes with respect to package loading and surface storage time;
* make discussions about the relationship between fuel cycle, disposal concept, and waste management strategy.

## 2. Fuel cycle wastes

The fuel cycles and the properties of their waste streams are obtained from [E. Hardin et al.](https://energy.sandia.gov/wp-content/gallery/uploads/116202.pdf) (2011). The mass of waste produced from each fuel cycle *at steady state operation* is normalized to unit of electricity produced. To do this, we can utilize the burnup of the used fuels discharged from the reactors. For each fuel cycle, E. Hardin et al gives the discharge burnup for each used fuel. This can be leveraged to normalize the amount of waste produced in each fuel cycle to a common basis.

$$
\text{burnup} = \frac{\text{GWd}}{\text{t}}
$$

With the mass of the fuel required, we can determine based on additional data how many canisters will be required for disposal. For direct disposal of high-level wastes (HLW), it is assumed that each disposal package contains a single waste form. For direct disposal of used nuclear fuel (UNF), variable package loading is considered, ranging from a single assembly up to 12 assemblies per package. 

```python
pkg_loading = {}
pkg_loading['uox'] = [1, 2, 3, 4, 12] # asm./pkg
pkg_loading['mox'] = [1, 2, 3, 4, 12] # asm./pkg
pkg_loading['coex'] = [1] # wf/pkg
pkg_loading['nuex'] = [1] # wf/pkg
pkg_loading['ecc'] = [1] # wf/pkg
pkg_loading['ecm'] = [1] # wf/pkg
```

#### Once-through LWR fuel cycle

In the open fuel cycle, used uranium oxide (UOX) fuel from a pressurized light water reactor is directly disposed in a geologic repository. The fuel is assumed to have discharge burnup of 60 GWd/t. Assuming all assemblies have this discharge burnup and a thermal to electric efficiency of 33% for PWR, we can calculate the amount of fuel required to generate 100 GWe-y.

$$
M_{uox} = 100 GW_ey \times \frac{365 d}{1 y} \times \frac{1 GW_{th}}{0.33 GW_e} \times \frac{t}{60 GWd}
$$

Assuming that each assembly has a mass of approximately 460 kg of uranium [2], we can estimate the number of packages required for disposal of the UOX LWR UNF based on the number of packages.

```python
# once through fuel cycle
burnup = 60.0
eff_lwr = 0.33
m_ass = 0.460 # tU/assembly [2]
# total mass of unf
M = 100*365/eff_lwr/burnup
print(str(round(M,1))+' t UOX LWR fuel required to generate 100 GWe-y')
# number of uox packages
pkgs_uox = {}
for n in pkg_loading['uox']:
    pkgs_uox[n] = int(np.ceil(M/m_ass/n))
    print(str(pkgs_uox[n])+' '+str(n)+'-assembly packages required')
```

#### Limited-recycle LWR-LWR fuel cycle

LWR UOX used fuel (50 GWd/t) is reprocessed via the co-extraction (COEX) method to recover uranium and plutonium for mixed-oxide (MOX) fuel. The MOX fuel is burned once in an LWR (50 GWd/t) and sent for direct disposal. The waste forms produced are the COEX HLW glass and used MOX fuel.

The partially closed fuel cycle is a bit more complicated than the once-through fuel cycle because it is made up of two stages, both producing power. The report specifies that each MT of reprocessed LWR fuel allows fabrication of 108.9 kg of MOX fuel. This is a linear problem with four unknowns and four equations. 

Unknowns: 

1. $E_{uox}$, the amount of electricity generated from the UOX-fueled PWR
2. $E_{mox}$, the amount of electricity generated from the MOX-fueled PWR
3. $M_{uox}$, the amount of UOX fuel required
4. $M_{mox}$, the amount of MOX fuel required

Equations:

1. $M_{uox} = E_{uox} [GW_ey] \times \frac{365 d}{1 y} \times \frac{1 GW_{th}}{0.33 GW_e} \times \frac{t}{50 GWd}$
2. $M_{mox} = E_{mox} [GW_ey] \times \frac{365 d}{1 y} \times \frac{1 GW_{th}}{0.33 GW_e} \times \frac{t}{50 GWd}$
3. $M_{mox} = 0.1089M_{uox}$
4. $E_{uox}+E_{mox} = 100 GW_ey$

```python
burnup_uox = 50.0
burnup_mox = 50.0
A = np.array([[1, 0, -365/(eff_lwr*burnup_uox), 0],
               [0, 1, 0, -365/(eff_lwr*burnup_mox)],
               [0.1089, -1, 0, 0],
               [0, 0, 1, 1]])
b = np.array([0, 0, 0, 100])
```

```python
ans = np.linalg.solve(A,b.transpose())
print(str(round(ans[0],1))+' t UOX LWR fuel required for partially closed fuel cycle (100 GWe-y total)')
print(str(round(ans[1],1))+' t MOX LWR fuel required for partially closed fuel cycle (100 GWe-y total)')
```

Note here that even though a second fuel cycle stage is utilized to achieve the total of 100 GWe-y generation, more UOX fuel is required than in the once through cycle. The reason for this is that less energy is extracted from the fuel - in the partially closed fuel cycle, the UOX burnup is lower than in the once-through fuel cycle, in part to preserve plutonium for reuse in the MOX fuel.

In the COEX process, the uranium and plutonium in the UOX fuel is removed, leaving behind fission products that are disposed in a HLW glass waste form. 537.5 kg of glass is formed for every 1 ton of reprocessed LWR UOX fuel; each canister contains 2900 kg glass. It is assumed that the excess uranium *not* recovered for reuse in the MOX fuel is not sent to the geologic repository but is treated as low-level waste.

We approximate the mass of the MOX fuel assembly to be equal to that of the UOX assembly.

```python
# coex hlw
pkgs_coex = {}
for n in pkg_loading['coex']:
    pkgs_coex[n] = int(np.ceil(ans[0]*0.5375/2.9/n))
    print(str(pkgs_coex[n])+' 1-unit packages required for COEX glass')
# mox unf
m_ass = 0.460 # tU/assembly [2]
pkgs_mox = {}
for n in pkg_loading['mox']:
    pkgs_mox[n] = int(np.ceil(ans[1]/m_ass/n))
    print(str(pkgs_mox[n])+' '+str(n)+'-assembly packages required for MOX UNF')
```

#### Continuous-recycle LWR-SFR fuel cycle

In this fuel cycle, other than losses, no actinides are sent to the repository. LWR UOX is reprocessed via the new-extraction (NUEX) method to recover uranium and transuranic (TRU) elements for recycle as metal fuel in a sodium fast reactor (SFR). The SFR used fuel is recycled using electrochemical processing to separate recycled actinides and fission products. The waste forms produced are the NUEX HLW glass and the electrochemical processing ceramic and metal HLWs.

Again, we have 4 equations and 4 unknowns. The TRU enrichment in the LWR used fuel is equal to the sum of the Pu and MA enrichments, which are 1.17 wt.% and 0.14 wt.%, respectively. The SFR requires 0.75 t TRU makeup, of which some must come from LWR UNF. 5.78 t LWR UNF are required to supply TRU to the SFR per year. Annually, the SFR requires 3.8 t of new fuel (HM (Pu+TRU): 3.55 t; TRU: 0.75 t; Zr: 0.25 t). Therefore, we can say that 5.78 t LWR UNF makes possible the fabrication of 3.8 t of new SFR fuel: $M_{sfr}=0.657M_{uox}$ [1]. 

Note the higher burnup and thermal efficiency in the SFR. The thermal efficiency is higher because the reactor operates at higher temperatures than LWR.

Unknowns: 

1. $E_{uox}$, the amount of electricity generated from the UOX-fueled PWR
2. $E_{sfr}$, the amount of electricity generated from the SFR
3. $M_{uox}$, the amount of UOX fuel required
4. $M_{sfr}$, the amount of SFR fuel required

Equations:

1. $M_{uox} = E_{uox} [GW_ey] \times \frac{365 d}{1 y} \times \frac{1 GW_{th}}{0.33 GW_e} \times \frac{t}{50 GWd}$
2. $M_{sfr} = E_{sfr} [GW_ey] \times \frac{365 d}{1 y} \times \frac{1 GW_{th}}{0.40 GW_e} \times \frac{t}{99.6 GWd}$
3. $M_{sfr} = 0.657M_{uox}$
4. $E_{uox}+E_{sfr} = 100 GW_ey$

```python
burnup_uox = 50.0
burnup_sfr = 99.6
eff_sfr = 0.40
A = np.array([[1, 0, -365/(eff_lwr*burnup_uox), 0],
               [0, 1, 0, -365/(eff_sfr*burnup_sfr)],
               [0.657, -1, 0, 0],
               [0, 0, 1, 1]])
b = np.array([0, 0, 0, 100])
```

```python
ans = np.linalg.solve(A,b.transpose())
print(str(round(ans[0],1))+' t UOX LWR fuel required for fully closed fuel cycle (100 GWe-y total)')
print(str(round(ans[1],1))+' t metal SFR fuel required for fully closed fuel cycle (100 GWe-y total)')
```

The UOX used fuel is reprocessed in the NUEX process, forming 309.2 kg of glass for every ton of fuel reprocessed. Electrochemical reprocessing of the SFR metal fuel produces 2,641 kg of ceramic waste and 4,403 kg of metal waste for every ton of fuel reprocessed. The canisters for the NUEX and E-chem ceramic HLW each contain 2900 kg of waste. The canisters for the E-chem metal waste contain 3600 kg.

```python
# nuex hlw
pkgs_nuex = {}
for n in pkg_loading['nuex']:
    pkgs_nuex[n] = int(np.ceil(ans[0]*0.3092/2.9/n))
    print(str(pkgs_nuex[n])+' 1-unit packages required for NUEX glass')
# ec-ceramic hlw
pkgs_ecc = {}
for n in pkg_loading['ecc']:
    pkgs_ecc[n] = int(np.ceil(ans[1]*2.641/2.9/n))
    print(str(pkgs_ecc[n])+' 1-unit packages required for echem ceramic waste')
# ec-metal hlw
pkgs_ecm = {}
for n in pkg_loading['ecc']:
    pkgs_ecm[n] = int(np.ceil(ans[1]*4.403/3.6/n))
    print(str(pkgs_ecm[n])+' 1-unit packages required for echem metal waste')
```

```python
waste = {}
waste['once-through'] = {'uox':pkgs_uox}
waste['limited-recycle'] = {'mox':pkgs_mox, 'coex':pkgs_coex}
waste['continuous-recycle'] = {'nuex':pkgs_nuex, 'ecc':pkgs_ecc, 'ecm':pkgs_ecm}
```

## 3. Determination of repository footprint

Waste heat generation will have various effects on the size of the repository. First, for combinations of fuel cycle and geology where the waste does not breach thermal constraints, the spacing between the heat-emitting canisters is determined by the decay heat of the waste, enabling the possibility of reducing repository size by decreasing thermal output. Second, certain disposal combinations of fuel cycle and geology will not be possible because the thermal output of the waste breaches thermal constraints.

Note: This calculation assumes that all canisters are emplaced at the same time and after the same amount of storage time. One option not considered in this work for improving repository utilization is "staggered loading" in which alternating drifts are filled over a period of years - later, after heat has been dissipated from those drifts, the adjacent drifts are loaded. 

In an adjacent notebook, the disposal area-per-canister (APC) can be calculated for each waste form and each repository geology at a specified surface storage time. The storage time adds an extra dimension to the analysis. The longer waste is left on the surface, the cooler it is when placed in the repository. For some storage times, no possible disposal of a waste form may be possible in a certain rock. For some rocks with stricter thermal constraints, disposal of a waste form may be impossible altogether.

```python
# data directory
app_data_dir = './footprint_data/area_per_package/'
st_data_dir = './footprint_data/storage_times/'
```

```python
# function to read in data
# structure of data: storage time , package loading , area per package
def import_app_data(path):
    df = pd.read_csv(path, names=['st', 'n', 'app'])
    return(df)
```

```python
 # Dictionary structure
# area per package
# -- waste form
# ---- pkg loading
# ------- surface storage time vector
# ------- area per package vector
area_per_pkg = {}
for r in ['granite', 'clay', 'salt']:
    area_per_pkg[r] = {}
    for w in ['uox', 'mox', 'coex', 'nuex', 'ecc', 'ecm']:
        area_per_pkg[r][w] = {}
        data_file = r+'_'+w+'.dat'
        x = import_app_data(os.path.join(app_data_dir, data_file))
        for n in list(set(x['n'])): # pkg loading
            area_per_pkg[r][w][n] = {}
            # subset the data frame to get only the requested pkg loading
            temp = x[x['n']==n].sort_values('st')
            # where app==0.0, no disposal is possible
            temp = temp[temp.app != 0.0]
            area_per_pkg[r][w][n]['st'] = temp['st']
            area_per_pkg[r][w][n]['app'] = temp['app']        
```

Now, for the surface storage time data

```python
# function to read in data
# structure of data: package loading, required st, maximum st
def import_st_data(path):
    df = pd.read_csv(path, names=['n', 'min', 'max'])
    return(df)
```

```python
rock = ['granite', 'clay', 'salt']
wasteforms = ['uox', 'mox', 'coex', 'nuex', 'ecc', 'ecm']
# Dictionary structure
# area per package
# -- waste form
# ---- pkg loading
# ------- min surface storage time
# ------- max surface storage time
surface_storage_times = {}
for r in rock:
    surface_storage_times[r] = {}
    for w in wasteforms:
        surface_storage_times[r][w] = {}
        data_file = r+'_'+w+'.st'
        x = import_st_data(os.path.join(st_data_dir, data_file))
        for n in list(set(x['n'])): # pkg loading
            surface_storage_times[r][w][n] = [float(x[x['n']==n]['min']), float(x[x['n']==n]['max'])]            
```

This section shows the area required per package for the waste streams produced in each fuel cycle as a function of disposal concept, surface storage time, and package loading.

```python
def get_letter(number):
    return(chr(number+65).lower())
```

#### Once-through fuel cycle

```python
# PLOT SETTINGS
# colors differentiate among package loadings
colors = {1:'k', 2:'b', 3:'g', 4:'y', 12:'r'}
x_max = 255.0
y_max = 520.0 
# plot everything
f, axarr = plt.subplots(3, sharex=True, figsize=(10, 12))
big = f.add_subplot(111, frameon=False)
big.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
big.grid(False)
# Set common labels
big.set_xlabel('Surface storage time (y)', fontsize=15)
big.set_ylabel(r'Area per package (m$^2$)', fontsize=15)
for r in range(0, len(area_per_pkg.keys())): # rock type
    rock = area_per_pkg.keys()[r]
    for pl in area_per_pkg[rock]['uox']: # package loading
        try:
            x = area_per_pkg[rock]['uox'][pl]['st']
            y = area_per_pkg[rock]['uox'][pl]['app']
            axarr[r].plot(x, y, marker='o', linestyle='-', color=colors[pl], 
                          label=str(pl)+'-assembly package', markersize=8)
        except:
            print('not possible to dispose of '+str(pl)+'-assembly '+
                  'package in a '+rock+' repository.')
    # set x and y limits
    axarr[r].set_xlim([1, x_max])
    axarr[r].set_ylim([0, y_max])
    # format tick value font
    for tick in axarr[r].yaxis.get_major_ticks():
        tick.label.set_fontsize(11)
    for tick in axarr[r].xaxis.get_major_ticks():
        tick.label.set_fontsize(11)
    # annotate figure
    axarr[r].text(0.02, 0.95, get_letter(r)+') '+rock.title(),
        verticalalignment='top', horizontalalignment='left',
        transform=axarr[r].transAxes,
        color='black', fontsize=15)
    # legend, xticks
    handles, labels = axarr[r].get_legend_handles_labels()
    if(r==0):
        axarr[r].xaxis.set_ticks(np.arange(0, x_max+10.0, 20.0))
    if(r==1):axarr[r].legend(handles, labels, loc='upper left', bbox_to_anchor=(0.65, 0.95))
f.subplots_adjust(hspace=0.05)
```

#### Repository area

In this section, I've written functions in order to sum the product of APP and package count to determine repository area for each fuel cycle. The result is a set of figures for the repository footprint (y-axis) as a function of:
* fuel cycle (parameterized on plot; distinguished by line style and color)
* repository type (subfigures)
* surface storage time (x-axis)
* package loading (parameterized on plot; distinguished by marker)

The surface storage time is limited to 250 years.

```python
def ot_area(rock, app_dict, waste_dict, uox_pkg_loading, st_max):
    "butthole"
    
    n_pkg_uox = waste_dict['once-through']['uox'][uox_pkg_loading]
    st_uox = app_dict[rock]['uox'][uox_pkg_loading]['st'][app_dict[rock]['uox'][uox_pkg_loading]['st']<=st_max]
    a_uox = ((app_dict[rock]['uox'][uox_pkg_loading]['app'][app_dict[rock]['uox'][uox_pkg_loading]['st']<=st_max])*n_pkg_uox)/1000.0/1000.0
    return(np.array(st_uox), np.array(a_uox))
```

```python
def lr_area(rock, app_dict, waste_dict, mox_pkg_loading, st_max):
    "butthole 2"
    
    n_pkg_mox = waste_dict['limited-recycle']['mox'][mox_pkg_loading]
    n_pkg_coex = waste_dict['limited-recycle']['coex'][1]
    st_coex = app_dict[rock]['coex'][1]['st'][app_dict[rock]['coex'][1]['st']<=st_max]
    a_coex = (app_dict[rock]['coex'][1]['app'][app_dict[rock]['coex'][1]['st']<=st_max])*n_pkg_coex
    st_mox = app_dict[rock]['mox'][mox_pkg_loading]['st'][app_dict[rock]['mox'][mox_pkg_loading]['st']<=st_max]
    a_mox = (app_dict[rock]['mox'][mox_pkg_loading]['app'][app_dict[rock]['mox'][mox_pkg_loading]['st']<=st_max])*n_pkg_mox
    st = np.array([x for x in [st_coex, st_mox] if len(x)==min([len(st_coex), len(st_mox)])][0])
    area = np.zeros(len(st))
    for i in xrange(0, len(st)):
        time = st[i]
        area[i] = (float(a_coex[st_coex==time])+float(a_mox[st_mox==time]))/1000.0/1000.0
    return(st, area)
```

```python
def cr_area(rock, app_dict, waste_dict, st_max):
    "butthole"
    
    n_pkg_nuex = waste_dict['continuous-recycle']['nuex'][1]
    n_pkg_ecc = waste_dict['continuous-recycle']['ecc'][1]
    n_pkg_ecm = waste_dict['continuous-recycle']['ecm'][1]
    # NUEX
    st_nuex = app_dict[rock]['nuex'][1]['st'][app_dict[rock]['nuex'][1]['st']<=st_max]
    a_nuex = (app_dict[rock]['nuex'][1]['app'][app_dict[rock]['nuex'][1]['st']<=st_max])*n_pkg_nuex
    # ECC
    # NUEX
    st_ecc = app_dict[rock]['ecc'][1]['st'][app_dict[rock]['ecc'][1]['st']<=st_max]
    a_ecc = (app_dict[rock]['ecc'][1]['app'][app_dict[rock]['ecc'][1]['st']<=st_max])*n_pkg_ecc
    # ECM
    st_ecm = app_dict[rock]['ecm'][1]['st'][app_dict[rock]['ecm'][1]['st']<=st_max]
    a_ecm = (app_dict[rock]['ecm'][1]['app'][app_dict[rock]['ecm'][1]['st']<=st_max])*n_pkg_ecm
    # consolidate
    st = np.array([x for x in [st_nuex,st_ecc,st_ecm] if len(x)==min([len(st_nuex),len(st_ecc),len(st_ecm)])][0])
    area = np.zeros(len(st))
    for i in xrange(0, len(st)):
        time = st[i]
        area[i] = (float(a_nuex[st_nuex==time])+float(a_ecc[st_ecc==time])+float(a_ecm[st_ecm==time]))/1000.0/1000.0
    return(st, area)
```

```python
def determine_area(rock, fuelcycle, app_dict, waste_dict, unf_pkg_loading=1.0, st_max=250.0):
    if(fuelcycle=='once-through'):
        return(ot_area(rock, app_dict, waste_dict, unf_pkg_loading, st_max))
    elif(fuelcycle=='limited-recycle'):
        return(lr_area(rock, app_dict, waste_dict, unf_pkg_loading, st_max))
    elif(fuelcycle=='continuous-recycle'):
        return(cr_area(rock, app_dict, waste_dict, st_max))
    else:
        return
```

```python
def get_minimum_area(rock, fuelcycle, area_dict, st):
    options = []
    for pl in area_dict[rock][fuelcycle].keys():
        if('st' in area_dict[rock][fuelcycle][pl].keys() and
           st in area_dict[rock][fuelcycle][pl]['st']):
            idx = np.where(area_dict[rock][fuelcycle][pl]['st']==st)[0][0]
            options.append(area_dict[rock][fuelcycle][pl]['area'][idx])
    try:
        return(min(options))
    except ValueError:
        return(0.0)
```

```python
def get_ratio_with_cr(area_dict, reference='once-through'):
    ratio = {}
    # rock; fuel cycle; st, ratio
    for rock in area_dict.keys():    
        ratio[rock] = {}
        for fuelcycle in area_dict[rock].keys():
            if(fuelcycle==reference):
                continue
            else:
                ratio[rock][fuelcycle] = {}
                ratio[rock][fuelcycle]['st'] = []
                ratio[rock][fuelcycle]['ratio'] = []
                for time in area_dict[rock][reference][1]['st']:
                    numerator = get_minimum_area(rock, fuelcycle, area_dict, time)
                    if(numerator == 0.0):
                        continue
                    else:
                        ratio[rock][fuelcycle]['st'].append(time)
                        denominator = get_minimum_area(rock, reference, area_dict, time)
                        ratio[rock][fuelcycle]['ratio'].append(numerator/denominator)
    return(ratio)
```
