In the repository footprint model the heat sources are moved closer together, challenging a few of the assumptions involved in the original model.

1. The time-dependent contribution from different sources in the array is calculated as conduction through solid, infinite host rock. As sources are moved closer to the evaluation point, the relative amount of host rock (as opposed to EBS materials) decreases. What happens when a greater share of the heat transfer takes place through more resistive material, and can we account for it?

2. The time-dependent contributions of heat from those sources are calculated independently. However, as they are moved closer together, the overall density of heat near the evaluation point is increased. Is it reasonable (or conservative) to calculate the contribution of heat from independent, individual sources?

In this notebook, data produced by COMSOL is used to assess these assumptions. The first section parses data from a model for a single heat source is evaluated to test the effect of a bentonite layer on the temperature distribution in an semi-infinite medium. In the second section, the data is for a 3x3 array heat sources embedded in a semi-infinite medium and the peak temperature at the evaluation point is assessed with and without bentonite buffer material for different spacing between the packages.

In both models, the package and the bentonite buffer are assumed to be spherical. The bulk medium is granite and is modeled as a large sphere with a 0K fixed-temperature boundary condition.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. The effect of bentonite layer for a single heat source

```python
# rock = open('./comsol/1d_rock.txt').read().splitlines()
# comp = open('./comsol/1d_composite.txt').read().splitlines()
rock = open('./comsol/recomsoldata/single_nobuffer_line.txt').read().splitlines()
comp = open('./comsol/recomsoldata/single_buffer_line.txt').read().splitlines()
data = [rock, comp]
labels = ['rock', 'comp']
comsol={}
for i in xrange(0, len(data)):
    txtlines=data[i]
    comsol[labels[i]]={}
    list_of_lines = []
    #line_idx=0
    for line_idx in xrange(9, len(txtlines)):    
        line = ['']
        col_idx = 0
        for char in xrange(0, len(txtlines[line_idx])):
            #print(str(char_idx)+' '+str(col_idx))
            if(txtlines[line_idx][char]!=' '):
                line[col_idx]+=txtlines[line_idx][char]
            else:
                upnext = txtlines[line_idx][char+1]
                if(upnext!=' '):
                    line.append('')
                    col_idx+=1
        list_of_lines.append(line)
    for l in xrange(0, len(list_of_lines)):
        comsol[labels[i]][list_of_lines[l][0]] = list_of_lines[l][3:]
```

```python
# Change values to float
for x in comsol.keys():
    for d in comsol[x].keys():
        for i in xrange(0,len(comsol[x][d])):
            T = float(comsol[x][d][i])
            if(T<0.0):
                T=0.0
            comsol[x][d][i] = T
```

```python
# Change keys to float
for x in comsol.keys():
    comsol[x] = dict((float(k),v) for k,v in comsol[x].iteritems())
```

```python
plt.plot(np.sort(comsol['rock'].keys()), [max(comsol['rock'][d]) for d in np.sort(comsol['rock'].keys())], 'bo--', label='Rock')
plt.plot(np.sort(comsol['comp'].keys()), [max(comsol['comp'][d]) for d in np.sort(comsol['comp'].keys())], 'rs-', label='EBS-rock composite')
plt.plot([0.85, 0.85], [0, 500], 'k--')
plt.plot([1.2, 1.2], [0,500], 'k--')
#plt.ylim(ymax=100)
#plt.ylim(ymax=20)
plt.xlim([0,10])
plt.xlabel('Distance between spherical source center and evaluation point (m)')
plt.ylabel('Peak temp increase (C)')
#plt.legend(loc=(1.1, 0.35))
plt.legend(loc='upper right')#loc=(1.1, 0.35), bbox_to_anchor=(1.03, 0.4))
plt.savefig('./2019-05-06_peaktemp.png', dpi=200)
plt.show()
```

```python
labels = ['buffer', 'nobuffer']
single_src_evalpt = {}
for b in labels:
    datafile = './comsol/recomsoldata/single_'+b+'_pt.txt'
    temp = open(datafile).read().splitlines()
    line = ['']
    col_idx=0
    for char in xrange(0, len(temp[9])):
        if(temp[9][char]!=' '):
            line[col_idx]+=temp[9][char]
        else:
            upnext = temp[9][char+1]
            if(upnext!=' '):
                line.append('')
                col_idx+=1
    single_src_evalpt[b] = [float(x) for x in line[3:]]
```

```python
plt.plot(np.arange(0, 100+0.5, 0.5), single_src_evalpt['buffer'], 'r-', label='EBS-rock composite')
plt.plot(np.arange(0, 100+0.5, 0.5), single_src_evalpt['nobuffer'], 'b-', label='Rock only')
plt.ylim(ymin=0)
plt.xlim([0,10])
plt.xlabel('Time (y)')
plt.ylabel('Temp increase (C)')
plt.legend(loc='upper right')#loc=(1.1, 0.35), bbox_to_anchor=(1.03, 0.4))
# plt.savefig('./2019-04-29_peaktemp.png', dpi=200)
plt.show()
```

## 2. Effect of bentonite layer with multiple sources at different distances

```python
distances = [2.4, 4, 6, 8, 10, 12, 14, 16]
distances = [2.4, 4, 8, 12, 16]
T = {}
T['buffer'] = {}
T['nobuffer'] = {}
times = np.arange(0, 100.5, 0.5)
```

```python
for b in T.keys():
    for d in distances:
        #datafile = './comsol/comsoldata/pt_'+b+'_d='+str(d)+'.txt'
        datafile = './comsol/comsoldata_fine/pt_'+b+'_d='+str(d)+'.txt'
        #print datafile
        temp = open(datafile).read().splitlines()
        line = ['']
        col_idx=0
        for char in xrange(0, len(temp[9])):
            if(temp[9][char]!=' '):
                line[col_idx]+=temp[9][char]
            else:
                upnext = temp[9][char+1]
                if(upnext!=' '):
                    line.append('')
                    col_idx+=1
        T[b][d] = [float(x) for x in line[3:]]
```

```python
c = {2.4:'k', 4:'r', 6:'b', 8:'g', 10:'m', 12:'y', 14:'c', 16:'k'}
l = {'buffer':'-', 'nobuffer':'--'}
for d in distances:
    for b in T:
        if(b=='nobuffer'):
            lab='no buffer'
        else:
            lab=b
        plt.plot(times, T[b][d], color=c[d], linestyle=l[b], label='d='+str(d)+'; '+lab)
plt.xlabel('Time (y)')
plt.ylabel('Temperature increase (C)')
plt.ylim([0, 65])
# plt.xlim([0,10])
plt.legend(loc=(1.05, 0.08))
plt.savefig('./2019-05-07_evaluation_pt', dpi=200,  bbox_inches='tight')
```

```python
print('Peak temperature at evaluation point')
print('Distance'+'\t'+'Buffer'+'\t\t'+'No buffer')
for d in distances:
    print(str(d)+'\t\t'+str(round(max(T['buffer'][d]),3))+'\t\t'+str(round(max(T['nobuffer'][d]),3)))
```

```python
difference = []
for d in distances:
    difference.append(max(T['nobuffer'][d])-max(T['buffer'][d]))
```

```python
plt.plot(distances, difference, 'bo')
plt.xlabel('Distance between source centers (m)')
plt.ylabel(r'max($T_{rock}$)-max($T_{buffer}$) (C)')
plt.xlim([0, 17])
plt.savefig('./2019-05-07_diff.png', dpi=200)
```

```python
plt.plot(distances, [times[T['buffer'][d].index(max(T['buffer'][d]))] for d in distances], 'rs-', label='buffer')
plt.plot(distances, [times[T['nobuffer'][d].index(max(T['nobuffer'][d]))] for d in distances], 'bo-', label='nobuffer')
plt.xlabel('Distance between source centers')
plt.ylabel('Time of peak temperature increase (y)')
plt.legend(loc='lower right')
plt.xlim([0, 17])
plt.ylim([3,7])
plt.show()
```
