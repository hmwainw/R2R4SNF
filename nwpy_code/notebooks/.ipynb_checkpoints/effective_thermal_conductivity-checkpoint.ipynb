Many assumptions were made in the use of analytical solutions to evaluate the temperature constraint in a repository containing heat-generating waste. These assumptions are justified in previous works and are generally valid for the cases tested in those works. Most notably, in the repository designs tested, the heat sources are located relatively far apart.

In the repository footprint model the heat sources are moved closer together, challenging a few of the assumptions involved in the original model.

1. The time-dependent contribution from different sources in the array is calculated as conduction through solid, infinite host rock. As sources are moved closer to the evaluation point, the relative amount of host rock (as opposed to EBS materials) decreases. What happens when a greater share of the heat transfer takes place through more resistive material, and can we account for it?

2. The time-dependent contributions of heat from those sources are calculated independently. However, as they are moved closer together, the overall density of heat near the evaluation point is increased. Is it reasonable (or conservative) to calculate the contribution of heat from independent, individual sources?

This notebook attempts to resolve the effects of the first of these questions. Analysis will be for the granite repository concept, because it features the greatest difference between EBS and host rock heat transfer properties. Therefore, if the assumptions hold, or if new methods can account for that difference, they should also be able to work for the other repository concepts.

```python
# Buffer: Dry bentonite
k_buffer = 0.6 # W/m-K
p_buffer = 2290.0 # W/m-K
cp_buffer = 650.0 # J/kg-K
a_buffer = k_buffer/p_buffer/cp_buffer # W/m-K
# Rock: Granite
k_rock = 2.5 # W/m-K
a_rock = 1.13e-6 # m^2/s
```

```python
# Print ratio
print('ratio of thermal conductivity (rock to buffer): '+str(round(k_rock/k_buffer,3)))
print('ratio of thermal diffusivity (rock to buffer): '+str(round(a_rock/a_buffer,3)))
```

Effective medium theory proposes the adjustment of thermal properties to account for inclusions in the medium. Modeling conduction through foams or suspensions requires accounting for conductivity of both the medium and the inclusions. A simple adjustment can be made using the following method, which relies on the volume fraction of inclusions in the matrix.

```python
def k_maxwell(km, k1, phi):
    denom = (k1+2*km)/(k1-km)-phi
    return(km*(1+(3*phi/denom)))
```

## 0. Analytical models for heat conduction

In this section, the analytical models for heat conduction from different sources are implemented.

```python
def gen_tpts(endtime=500.0):
    t = np.arange(-2.0, np.log10(endtime), 0.1)
    t = np.append(t, np.log10(endtime))
    return(10**t)
```

```python
def integrate(integrandfxn, target, arguments=()):
    """Integrate a function from 0 to some value using scipy.interpolate.quad function"""

    abserr = 1.0
    counter = 1
    # The number of integration subdivisions is automatically increased if scipy.integrate.quad
    # fails. To prevent warnings every time this occurs, all warnings are ignored; this may
    # be bad practice, suggestions welcome.
    warnings.filterwarnings("ignore")
    # Iterate to compute the integral
    while(abserr > 0.0001):
        x, abserr = quad(integrandfxn, 0, target, args=arguments, limit=50*counter)[0:2]
        counter=counter+1
        if(counter > 5):
            break
    # Reset warnings, check for convergence
    warnings.resetwarnings()
    if(abserr > 0.0001):# and abserr > tol):
        warnings.warn('Integral not converged after 250 subdivisions; abserr = '+str(abserr),
                      RuntimeWarning)
    # return evaluated integral
    return(x)
```

```python
def sphere(r_in, r_out, trgt, q, k, a, st=0.0):
    
    a *= 365*24*3600
    # time-dependent heat generation
    # integrand function
    def integrand(t,ri,ro):
        # q given in Km3/s
        #c = 1.0/8.0/np.pi/ro/ri/np.sqrt(np.pi*a*(trgt-t))
        # q given in W
        c = np.sqrt(a)/8.0/np.pi/ro/ri/k/np.sqrt(np.pi*(trgt-t))
        #denom = 8.0*(np.pi*a*(trgt-t))**1.5
        expterm1 = np.exp(-(ro-ri)**2/4.0/a/(trgt-t))
        expterm2 = np.exp(-(ro+ri)**2/4.0/a/(trgt-t))
        hterm = q(t+st)
        return(c*hterm*(expterm1-expterm2))
    # compute the integral
    return(integrate(integrand, trgt, arguments=(r_in,r_out)))
```

```python
def point(r, trgt, q, k, a, st=0.0):
    """"""
    a *= 365*24*3600
    # time-dependent heat generation
    # integrand function
    def integrand(t,dist):
        # q given in Km3/s
        #denom = 8.0*(a**1.5)*(np.pi**1.5)*((trgt-t)**1.5)
        # q given in W
        denom = 8.0*k*np.sqrt(a)*(np.pi**1.5)*((trgt-t)**1.5)
        expterm = np.exp(-(dist**2)/4.0/a/(trgt-t))
        hterm = q(t+st)
        return(hterm*expterm/denom)
    # compute the integral
    return(integrate(integrand, trgt, arguments=(r,)))
```

#### 1.0.3 Spatial limits

The two types of sources (point and spherical) have different $r_{min}$. The volume of bentonite is to be kept constant based on EBS designs. In the case of the point source, the heat will be released from the point into the center of a bentonite sphere. For the spherical source, its volume will be equal to that of the waste package, and it will emit heat into a spherical bentonite shell. Therefore, $r_{min}$ for the spherical source will be larger than that for the point source.

```python
d_pkg = 0.82 # m; 4-asm design
l_pkg = 5.0 # m
dr_buffer = 0.35 # bentonite
v_pkg = l_pkg*np.pi*(d_pkg/2)**2 # cylinder
# convert cylindrical volume into spherical radius
r_pkg = (0.75*v_pkg/np.pi)**(1/3.0) # m
print r_pkg
```

```python
# POINT SOURCE - neglect the package
# Calculate the volume of the bentonite shell; 
# the bentonite sphere will surround the point source
v_bent = np.pi*(r_pkg+dr_buffer)**3/0.75 - v_pkg
r_ebs_pt = (0.75*v_bent/np.pi)**(1/3.0)
print('Point source EBS radius: '+str(round(r_ebs_pt,3)))
```

```python
# SPHERICAL SOURCE - account for the package
r_ebs_sph = r_pkg+dr_buffer # m; outer radius
print('Spherical source EBS radius: '+str(round(r_ebs_sph,3)))
```

```python
# MAXIMUM RADIUS
# Assume that if assumptions worked for this distance
# in previous works, they'll work here too
r_max = 20.0 # m; value used in previous works
```

```python
def calculate_ebs_vf(r, r_ebs):
    v_ebs = np.pi*r_ebs**3/0.75
    v_tot = np.pi*r**3/0.75
    return(v_ebs/v_tot)
```

## Two-stage analytical conduction

1. Solve the heat transfer from a spherical source through an infinite EBS region: obtain T1
2. Obtain T1(r_ebs, t), the radius of the EBS boundary
3. Use the result from (2) as the heat flux for a spherical source into an infinite rock region: obtain T2

As long as the EBS region is more thermally resistive than the rock region, this should be conservative.

```python
delta = 0.01
T1 = np.zeros(len(t_pts)) # buffer
T2 = np.zeros(len(t_pts)) 
for i in xrange(0,len(t_pts)):
    T1[i] = sphere(r_pkg, r_ebs_sph-delta, t_pts[i], decayheat, k_buffer, a_buffer)
    T2[i] = sphere(r_pkg, r_ebs_sph+delta, t_pts[i], decayheat, k_buffer, a_buffer)
```

```python
# get the heat flux
q_bent2rock = np.zeros(len(t_pts))
for i in xrange(0, len(t_pts)):
    q_bent2rock[i] = -k_buffer*(T2[i]-T1[i]/(2*delta))
```

```python
plt.plot(t_pts, q_bent2rock)
plt.plot(t_pts, [decayheat(t) for t in t_pts])
```

```python
logq_b2r = np.log10(q_bent2rock)
q_b2r = interp1d(t_pts, logq_b2r, fill_value="extrapolate")
def heat_bent2rock(t):
    return(10.0**q_b2r(t))
```

```python
T3 = {}
for d in dist_sph[1:]:
    T3[d] = np.zeros(len(t_pts)) 
    for i in xrange(0,len(t_pts)):
        T3[d][i] = sphere(r_ebs_sph, d, t_pts[i], heat_bent2rock, k_rock, a_rock)
```

```python
T_qss = {}
for d in dist_sph[1:]:
    T_qss[d] = np.zeros(len(t_pts)) 
    for i in xrange(0,len(t_pts)):
        T_qss[d][i] = sphere(r_ebs_sph, d, t_pts[i], decayheat, k_rock, a_rock)
```
