```python
%matplotlib inline
import openmc
import math

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.ticker as tk 

import numpy as np
import seaborn as sns
from scipy.stats import pearsonr

pd.set_option('display.precision', 2)
```

```python
# Load the csv files obtained with ...

reactors=['Ref_PWR', 'Micro_PWR', 'HTGR_80', 'eVinci', 'USNC']
reactors_legend=reactors # if the legend is different from the name of the folders

# Return a list containing the dataframe with the results for each reactors
def load_results(reactors):
    results=[]
    for i in reactors:
        result=pd.read_csv('/home/cchristiaen/Etudes_OPENMC/ResultsOK/'+i+'/cooling'+i+'.csv', sep=',') # path to the csv files
        results.append(result)
    return results
```

```python
# Change the index of the csv file. The time at discharge is 0. 
def set_cooling_time(results):    
    for i in range (len(results)):
        discharge=results[i].iloc[1, 0]
        results[i]['cooling time']=(results[i].iloc[:, 0]-discharge)/365
        results[i].set_index('cooling time', inplace=True)
    return results
```

```python
# Load the results
results=load_results(reactors)
results=set_cooling_time(results)
```

```python
# Function to add value labels on the bar chart

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

# Summary of the results and comparison with the reference

```python
# Enable to select the nearest index to the time in the results
## for example to select the activity at 10y, return the index the closest to 10 y and the corresponding value

def select_time(result, time ):
    res=result.copy(deep=True)
    res['time']=res.index.tolist()
    idx= res['time'].sub(time).abs().idxmin()
    return idx
    
```

```python
# Create a dataframe with the parameters of interest and select the corresponding values for each reactor

comp=pd.DataFrame(columns=['SNF mass (t/GWe.y)','SNF volume(m3/GWe.y)', 'Decay heat 10y (W/GWe.year)','Decay heat 100y (W/GWe.year)','Activity 100y (Ci/GWe.year)','Activity 100000y (Ci/GWe.year)','Radiotoxicity 100000y (Sv/GWe.y)'])

for i in range (len(results)): # iterate on the list of reactors
    result=results[i].copy(deep=True)
    l=[]
    l.append(SNF_mass[i])
    l.append(SNF_vol[i])    
    idx=select_time(result, 10) # select the index for 10y
    l.append(result.loc[idx, 'Decay heat (W/GWe.y)'])
    idx=select_time(result, 100) # select the index for 100y

    l.append(result.loc[idx, 'Decay heat (W/GWe.y)'])
    l.append(result.loc[idx, 'Activity (Ci/GWe.y)'])

    idx=select_time(result, 100000) # Select the index for 100,000 years
    l.append(result.loc[idx, 'Activity (Ci/GWe.y)'])
    l.append(result.loc[idx, 'Radiotoxicity (Sv/GWe.y)'])
    comp.loc[reactors[i]]=l

        
    
```

```python
cm = sns.light_palette("green", as_cmap=True)

comp_styled=comp.style.format("{:.3}").background_gradient(cmap=cm)
comp_styled
```

```python
# Create a dataframe with the value of the ratio with the reference reactor

reference_reactor='Ref_PWR'
ratio=pd.DataFrame()
for i in comp.columns:
    ref_value=comp.loc[reference_reactor, i]
    ratio[i]=comp[i]/ref_value

ratio
```

```python
# Create a radar chart with the values of the ratio dataframe

# Set data
df = ratio.transpose()

category=df.index.tolist()
category=[*category, category[0]]

# For each reactor create a list with the values of the metrics
react=[]
for i in df.columns:
    l=df[i].tolist()
    l=[*l, l[0]]
    react.append(l)

# Set the location of the legend
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(category))

fig, ax = plt.subplots(figsize=(9,9), subplot_kw=dict(polar=True))

linestyle=['-', '-', '-.', '-.', '--', ':',':'] 
marker=["o", "x", "^", "^", "^", "s", "*", 'D']

# For each reactor, plot the line linking the parameters value
n=0
for i in react:
    ax.plot(label_loc, i, lw=2, ls=linestyle[n], marker=marker[n] )
    n+=1

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# align the labels
for label, angle in zip(ax.get_xticklabels(), label_loc):
    if angle in (0, np.pi):
        label.set_horizontalalignment('center')
    elif 0 < angle < np.pi:
        label.set_horizontalalignment('left')
    else:
        label.set_horizontalalignment('right')
ax.set_yscale('symlog', linthresh=0.1)
ax.set_ylim(0,20)


ax.legend(reactors_legend, loc='best', bbox_to_anchor=(1.3, 0.7), ncols=1)


labels_graph=['SNF Mass', 'SNF Volume', 'Decay heat 10y', 'Decay Heat 100y', 'Activity 100y', 'Radiotoxicity 100000y', 'Activity 100000y', '']


lines, labels = plt.thetagrids(np.degrees(label_loc), labels=labels_graph)



plt.show()
```

# Coefficient of determination

```python
# define the metrics to calculate the correlation coefficients
metrics=['Decay heat (W/GWe.y) 10y', 'Activity (Ci/GWe.y) 10y', 'Radiotoxicity (Sv/GWe.y) 10y','Decay heat (W/GWe.y) 100y', 'Activity (Ci/GWe.y) 100y', 'Radiotoxicity (Sv/GWe.y) 100y','Decay heat (W/GWe.y) 100000y', 'Activity (Ci/GWe.y) 100000y', 'Radiotoxicity (Sv/GWe.y) 100000y']
df_metrics=pd.DataFrame(columns=metrics)

# Select of the value for each reactor at the respective time
for i in range (len(results)):
    result=results[i].copy(deep=True)
    l=[]
    idx=select_time(result, 10)
    l.append(result.loc[idx, 'Decay heat (W/GWe.y)'])
    l.append(result.loc[idx, 'Activity (Ci/GWe.y)'])
    l.append(result.loc[idx, 'Radiotoxicity (Sv/GWe.y)'])
    idx=select_time(result, 100)
    l.append(result.loc[idx, 'Decay heat (W/GWe.y)'])
    l.append(result.loc[idx, 'Activity (Ci/GWe.y)'])
    l.append(result.loc[idx, 'Radiotoxicity (Sv/GWe.y)'])
    idx=select_time(result, 100000)
    l.append(result.loc[idx, 'Decay heat (W/GWe.y)'])
    l.append(result.loc[idx, 'Activity (Ci/GWe.y)'])
    l.append(result.loc[idx, 'Radiotoxicity (Sv/GWe.y)'])
    df_metrics.loc[reactors[i]]=l
df_m=df_metrics.reindex(['Activity (Ci/GWe.y) 10y','Activity (Ci/GWe.y) 100y', 'Activity (Ci/GWe.y) 100000y','Decay heat (W/GWe.y) 10y','Decay heat (W/GWe.y) 100y', 'Decay heat (W/GWe.y) 100000y', 'Radiotoxicity (Sv/GWe.y) 10y', 'Radiotoxicity (Sv/GWe.y) 100y', 'Radiotoxicity (Sv/GWe.y) 100000y'  ], axis=1)
df_m.columns=['Activity 10y','Activity 100y', 'Activity 100000y','Decay heat 10y','Decay heat 100y', 'Decay heat 100000y', 'Radiotoxicity 10y', 'Radiotoxicity 100y', 'Radiotoxicity 100000y']
```

```python
# Calculate the pearson coefficient

pearson=df_m.corr()
pearson_sqr=pearson*pearson
pearson_sqr
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(pearson_sqr, annot=True, cmap='Blues')
```
