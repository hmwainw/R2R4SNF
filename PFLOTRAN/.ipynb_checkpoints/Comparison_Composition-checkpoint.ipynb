```python
reactor_list=[ 'Ref_PWR', 'Micro_PWR','HTGR_60', 'HTGR_80', 'HTGR_100','USNC', 'eVinci', 'SFR']
discharge_time=[1273,560,1191,1589,1986, 3849, 1658, 9917]
dict_time=dict(zip(reactor_list, discharge_time))
print(dict_time)

    
```

```python
results_act=load_results(reactors, 'activity')
results_DH=load_results(reactors, 'decay_heat')
results_act_conv=load_results(reactors, 'activity_conv')
results_DH_conv=load_results(reactors, 'decay_heat_conv')
results_act_prop=load_results(reactors, 'activity_prop')
results_DH_prop=load_results(reactors, 'decay_heat_prop')
results_rad_prop=load_results(reactors, 'radiotoxicity_prop')
```

```python
def select_time(result, time_value, reactor):
    time_value=time_value*365+dict_time[reactor]
    res=result.transpose()
    time=res.index.tolist()
    time=[float(x) for x in time]
    res['time']=time
    idx= res['time'].sub(time_value).abs().idxmin()
    return idx
```

```python
def sort_df(result, number):
    df_sorted=pd.DataFrame()
    for column in result:
        df=result[column].sort_values(axis=0, ascending=False).head(number)
        df_sorted=pd.concat([df_sorted, df], axis=1)
    return df_sorted
    
```

```python
def sort_df_reactors(results, number):
    sort_results=[]
    for i in results:
        df=sort_df(i, number)
        sort_results.append(df)
    return sort_results
```

```python
number=5
time_value=100000
```

```python
# Plot the activity by nuclide

fig, ax = plt.subplots()

sort_results_act=sort_df_reactors(results_act, number)
df=pd.DataFrame()

for i in range (len(reactors)):
    idx=select_time(results_act[i], time_value, reactors[i])
    df_sorted=sort_results_act[i][idx].transpose().dropna()
    df=pd.concat([df, df_sorted], axis=1)
    

df.plot.bar(ax=ax, ylabel='Activity (Bq/g)', xlabel='Nuclide', title=str(number)+' isotopes contributing most to the activity for each reactor')
ax.legend(reactors)
df
```

```python
def prop_total(reactors, results, time, number):
    prop_reactors=pd.DataFrame()
    for i in range (len(reactors)):
        idx=select_time(results[i], time, reactors[i])
        df=results[i].loc[:, idx].copy(deep=True)
        prop_reactors[i]=df
    prop_tot=prop_reactors.sum(axis=1)/len(reactors)
    prop_tot=prop_tot.sort_values(axis=0, ascending=False)
    if (number!=None):
        prop_tot=prop_tot.head(number)
    return prop_tot
    
        
```
