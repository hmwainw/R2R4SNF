## Determination of repository footprint

Waste heat generation will have various effects on the size of the repository. First, for combinations of fuel cycle and geology where the waste does not breach thermal constraints, the spacing between the heat-emitting canisters is determined by the decay heat of the waste, enabling the possibility of reducing repository size by decreasing thermal output. Second, certain disposal combinations of fuel cycle and geology will not be possible because the thermal output of the waste breaches thermal constraints.

Note: This calculation assumes that all canisters are emplaced at the same time and after the same amount of storage time. One option not considered in this work for improving repository utilization is "staggered loading" in which alternating drifts are filled over a period of years - later, after heat has been dissipated from those drifts, the adjacent drifts are loaded. 

In an adjacent notebook, the disposal area-per-canister (APC) can be calculated for each waste form and each repository geology at a specified surface storage time. The storage time adds an extra dimension to the analysis. The longer waste is left on the surface, the cooler it is when placed in the repository. For some storage times, no possible disposal of a waste form may be possible in a certain rock. For some rocks with stricter thermal constraints, disposal of a waste form may be impossible altogether.

## Package count

```python
def make_pkg_dict(df, eg):
    
    d = {}
    d['hlw'] = np.array([0.0])
    d['snf'] = {}#np.array([])
    temp = copy.deepcopy(data_app[data_app['eg']==eg])
    temp = temp[temp['rock']=='clay']
    temp = temp[temp['st']==100.0]
    # SNF
    snf = temp[temp['strm'].str.contains('|'.join(['df', 'snf']))]
    d['snf'] = np.sort(list(snf.groupby('loading').sum()['pkgs']))[::-1]
    if(d['snf'].size==0):
        d['snf']=np.append(d['snf'], 0.0)
    # HLW
    hlw = temp[~temp['strm'].str.contains('|'.join(['df', 'snf']))]
    d['hlw'] = hlw['pkgs'].sum()
    return(d)
```

```python
# plot setup
labels = sorted(list(set(data_app['eg'])))
idx = np.arange(2, 2*len(labels)+1, 2)
colors = [(u'yellow', u'#FFFF00'), (u'gold', u'#FFD700'),
          (u'orange', u'#FFA500'), (u'darkorange', u'#FF8C00'),
          (u'red', u'#FF0000'), (u'darkred', u'#8B0000')]
w = 0.5
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(15)
p = []
for i in xrange(0, len(labels)):
    eg = labels[i]
    x = idx[i]
    butt = make_pkg_dict(data_app, eg)
    p.append(ax.bar(x, butt['hlw'], width=w, color='b', align='center'))
    for j in xrange(0, len(butt['snf'])):
        p.append(ax.bar(x, butt['snf'][j], width=w, color=colors[j], bottom=butt['hlw'], align='center'))
ax.set_ylabel(r'Number of packages per 100GWe-y', fontsize=20)
ax.set_yscale('log')
ax.set_xticks(idx)
ax.set_xticklabels([l.upper() for l in labels], rotation=90)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set_xlim(xmax=max(idx)+2)
fig.savefig('./packages.png', dpi=200)
fig.show()
```

```python
eg01_min = {}
for r in ['granite', 'clay', 'salt']:
    eg01_min[r] = {}
    eg01_min[r]['st'] = eg01[r][1]['st']
    eg01_min[r]['area'] = []
    for st in eg01_min[r]['st']:
        dick = []
        for pl in eg01[r].keys():
            try:
                idx = eg01[r][pl]['st'].index(st)
            except:
                continue
            else:
                dick.append(eg01[r][pl]['area'][idx])
        eg01_min[r]['area'].append(min(dick))
```

### All the Analysis Examples

```python
def get_chars(df, eg):
    """Get number of stages, waste streams, package loadings"""
    
    df = copy.deepcopy(df)
    df = df[df['eg']==eg]
    char = {}
    for n in set(df['stg']):
        char[n] = {}
        for wf in set(df[df['stg']==n]['strm']):
            char[n][wf]=list(set(df[(df['stg']==n) & (df['strm']==wf)]['loading']))
    return(char)
```

```python
def get_min_area(ogdf, eg, rock, st):
    """Get the minimum repository area for a requested surface storage time"""
    
    df = copy.deepcopy(ogdf)
    df = df[df['eg']==eg]
    df = df[df['st']==st]
    df = df[df['rock']==rock]
    df['area'] = df['pkgs']*df['app']
    chars = get_chars(df, eg)
    area = 0.0
    for n in chars.keys(): # stages
        for wf in chars[n]: # wasteforms
            temp = df[(df['stg']==n) & (df['strm']==wf)]
            temp = temp[temp['area']!=0]
            if(temp.empty):
                return(0.0)
            else:
                area+= min(temp['area'])/1e6
    return(area)
```

Comments about repository area:
* Package count is important in extreme cases (e.g. MSR) and affects the area of the generic salt repository to a greater extent due to the larger EBS region
* Disposal of SNF after recycling is generally unfavorable: many of the limited-recycle cases produce wastes that require larger repository areas or cannot be disposed of after 100.0 years of storage.
* Recycling all TRU rather than Pu improves repository area, especially for thermal reactors: comparison of EG19/EG20, EG21/EG22; less effective for fast reactors because TRU is not accumulated (EG23/EG24).
* For EG09 and EG11, the concentration of shorter-lived FP (Cs) in one waste form makes disposal more difficult


# Minimum surface storage time

```python
# IMPORT RST DATA
rst_datafile = './fces_footprint_data/fces_rst_data.csv'
columns = ['eg', 'stg', 'strm', 'loading', 'pkgs', 'rock', 'rst']
data_rst = pd.read_csv(rst_datafile, names=columns)
```

```python
def get_min_rst(ogdf, eg, rock):
    """Get the minimum surface storage time (fxn of package loading)"""
    
    df = copy.deepcopy(ogdf)
    df = df[df['eg']==eg]
    df = df[df['rock']==rock]
    chars = get_chars(df, eg)
    rst = 0.0
    for n in chars.keys(): # stages
        for wf in chars[n]: # wasteforms
            temp_rst = min(df[(df['stg']==n) & (df['strm']==wf)]['rst'])
            if(temp_rst > rst):
                rst = temp_rst
    return(rst)
```

# FCES Categorizations

```python
import fces
```

```python
def has_thermal(eg):
    if('thermal' in fces.spectrum[eg]):
        return True
    else:
        return False
```

```python
def where_thermal(eg):
    return(fces.spectrum[eg].index('thermal')+1)
```

```python
def has_fast(eg):
    if('fast' in fces.spectrum[eg]):
        return True
    else:
        return False
```

```python
def where_fast(eg):
    return(fces.spectrum[eg].index('fast')+1)
```

```python
def has_th(eg):
    if('th' in fces.feed[eg]):
        return(True)
    else:
        return(False)
```

```python
def high_bu(eg, lim=100.0):
    if(any([x > lim for x in fces.burnup[eg]])):
        return(True)
    else:
        return(False)
```

```python
def is_recovered(eg, r='pu'):
    try:
        assert(eg in fces.recovered.keys())
    except AssertionError:
        return(False)
    else:
        if(r in fces.recovered[eg]):
            return(True)
        else:
            return(False)
```
