```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
q = 10.0
k1 = 0.6
k2 = 2.5
# a = 4.0
b = 100.0
```

```python
def T1(x,a,b):
    c1 = -q/k1
    c2 = q*a/k1 - q*a/k2 + q*b/k2
    return(c1*x+c2)
```

```python
def T2(x,b):
    c3 = -q/k2
    return(c3*x-c3*b)
```

```python
def T(x, a, b):
    if(x<=a):
        T=T1(x,a,b)
    else:
        T=T2(x,b)
    return(T)
```

```python
x = np.arange(0, b+1, 1.0)
# temp_1 = [T(i, 0.0, b) for i in x]
# temp_2 = [T(i, 5.0, b+5.0) for i in x]
# temp_3 = [T(i, 10.0, b+10.0) for i in x]
# temp_4 = [T(i, 20.0, b+20.0) for i in x]

temp_1 = [T(i, 0.0, b) for i in x]
temp_2 = [T(i, 5.0, b) for i in x]
temp_3 = [T(i, 10.0, b) for i in x]
temp_4 = [T(i, 20.0, b) for i in x]
```

```python
plt.plot(x, temp_1, 'b-', label=r'x$_{int}$=0.0')
plt.plot(x, temp_2, 'r-', label=r'x$_{int}$=5.0')
plt.plot([5.0, 5.0], [0, max(temp_4+temp_1+temp_2+temp_3)+100], 'r--')
plt.plot(x, temp_3, 'g-', label=r'x$_{int}$=10.0')
plt.plot([10.0, 10.0], [0, max(temp_4+temp_1+temp_2+temp_3)+100], 'g--')
plt.plot(x, temp_4, 'y-', label=r'x$_{int}$=20.0')
plt.plot([20.0, 20.0], [0, max(temp_4+temp_1+temp_2+temp_3)+100], 'y--')
# plt.plot([8.0, 8.0], [0, max(temp_4+temp_1+temp_2+temp_3)+100], 'k--')
plt.ylim(ymax=max(temp_4+temp_1+temp_2+temp_3)+20)
plt.xlabel('x')
plt.ylabel('Temperature')
plt.legend()
plt.savefig('./1dtemp.png', dpi=200)
plt.show()
```
